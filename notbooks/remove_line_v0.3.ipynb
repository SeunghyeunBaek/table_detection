```python
# 이미지 출력 함수
def imshow(img):
    plt.figure(figsize=(20, 10))
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

```python
ret, thr1 = cv2.threshold(img, 190, 255, cv2.THRESH_BINARY)
```

```python
erotion = cv2.erode(thr1, np.ones((3, 3), np.uint8), iterations=1)
```

```python
kernel = np.ones((1, 30), np.uint8)
dilation = cv2.dilate(erotion, kernel, iterations=1)
```

```python
kernel = np.ones((32, 1), np.uint8)
dilation_col = cv2.dilate(erotion, kernel, iterations=1)
```

```python
plt.imshow(dilation_col)
```

```python
mod_img = copy.deepcopy(thr1)  # 이미지 복사
row_idx = [i for i in range(dilation.shape[0])]
col_idx = [i for i in range(dilation.shape[1])]
for row_id in row_idx:
    for col_id in col_idx:
        if dilation[row_id, col_id] < 255:
            mod_img[row_id, col_id] = 255
        if col_id > 11:
            if dilation[row_id, col_id - 11] < 255:
                mod_img[row_id, col_id] = 255
        if col_id < dilation.shape[1] - 11:
            if dilation[row_id, col_id + 11] < 255:
                mod_img[row_id, col_id] = 255
```

```python
row_idx = [i for i in range(dilation.shape[0])]
col_idx = [i for i in range(dilation.shape[1])]
for row_id in row_idx:
    for col_id in col_idx:
        if dilation_col[row_id, col_id] < 255:
            mod_img[row_id, col_id] = 255
        if row_id > 11:
            if dilation_col[row_id - 13, col_id] < 255:
                mod_img[row_id, col_id] = 255     
        if row_id < dilation_col.shape[0] - 13:
            if dilation_col[row_id + 13, col_id] < 255:
                mod_img[row_id, col_id] = 255                   
```
